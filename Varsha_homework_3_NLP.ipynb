Annotation guidelines:
    These are from the ACE Guidelines for Person:
    
3.1 Persons (PER) 
    Each distinct person or set of people mentioned in a document refers to an entity of type Person. For example, people may be specified by name (“John Smith”), occupation (“the butcher”), family relation (“dad”), pronoun (“he”), etc., or by some combination of these. Dead people and human remains are to be recorded as entities of type Person. So are fictional human characters appearing in movies, TV, books, plays, etc. 

There are a number of words that are ambiguous as to their referent. For example, nouns, which normally refer to animals or non-humans, can be used to describe people. If it is clear to the annotator that the noun refers to a person in a given context, it should be marked as a Person entity. 

Examples: 
He is [a real turkey]
[The political cat of the year]
She’s known as [the brain of the family]

```python
import json
from pprint import pprint


# method to read annotation file
def annotation_processor(annotation_file):
    annotation_array = []

    # here we need to be careful and process each line of the annotation file separately
    read_annotation = open(annotation_file)
    for line in read_annotation:
        data = json.loads(line)
        annotation_array.append(data)

    # here we return an array of the individual annotations
    return annotation_array
    
# calling the annotation processor function
annotation_processor('/Users/varsha/PythonWorkspace/homework_3/Annotated/annotated.json')
```

```python
def compare_articles(reference_annotations, my_annotations):              
    matches = 0
    nonmatches = 0
    for articles in range(len(reference_annotations)):                          # Loops on the number of Articles  
        if reference_annotations[articles] == my_annotations[articles]:   # Matches on the entire Article
            matches += 1
        else:
            nonmatches += 1
    return matches, nonmatches
```

```python

### Compare at Annotation Level
def compare_annotations_prelim(reference_annotations, personal_annotations):    # Loops on the Annotations themselves
    ref_points = [(reference_annotations[i]['points'][0]['start'],reference_annotations[i]['points'][0]['end']) for i in range(len(reference_annotations))]
    per_points = [(personal_annotations[i]['points'][0]['start'],personal_annotations[i]['points'][0]['end']) for i in range(len(personal_annotations))]
                                                                                # Finds the Start and End points for every Annotation in every Article
    num_matches = len(set(ref_points) & set(per_points))                        # Interection of two lists, to compare same Start and End points in both files
    num_non_matches = len(set(ref_points) | set(per_points)) - num_matches      # Union of the two lists, to compare Start and End points that only occur in one file
    return num_matches, num_non_matches
```

```python

def compare_annotations(reference_annotation_array, personal_annotation_array):
    num_matches = 0
    num_nonmatches = 0
    for j in reference_annotation_array:                                        #
        for k in personal_annotation_array:
            if (j["content"] == k["content"]):                                  # Ensures the same articles are being compared
                temp_num_matches, temp_num_nonmatches = compare_annotations_prelim(j["annotation"], k["annotation"])  # 
                #print(temp_num_matches, temp_num_non_matches)
                num_matches += temp_num_matches                                 # Implement the sum of the temp_num_matches to the num_matches
                num_nonmatches += temp_num_nonmatches
    return num_matches, num_nonmatches
```

Step 3:
With the numbers now available, try to determine the Cohen's Kappa for this dataset. (You can calculate this by hand if you prefer). Assume that the Probability of random agreement is 0.3.
Note, with this the proportionate agreement would be the number of matches divided by the number of matches + the number of non-matches.
The rest of the formula should be as discussed in class.
Report the numbers you calculate.

```python

def cohens_kappa(article_matches, article_nonmatches):                          # Function to calculate Cohen's Kappa
    observed_agreement = article_matches / (article_matches+article_nonmatches) # Calculates Observed Agreement 
    chance_agreement = 0.3                                                      # Pre-defined Chance Agreement
    kappa =  observed_agreement - chance_agreement/(1-chance_agreement)         # Cohen's Kappa Formula
    return observed_agreement, chance_agreement, kappa
```
